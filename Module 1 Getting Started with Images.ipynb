## Split the Image

```python
img = cv2.imread("New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)
b,g,r = cv2.split(img)
# adapter the shot, log begin loger
# 我靠，这里好像优先级不高啊。
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r, cmap="gray");plt.title("Red Channel");
plt.subplot(142);plt.imshow(g, cmap="gray");plt.title("Green Channel");
plt.subplot(143);plt.imshow(b, cmap="gray");plt.title("Blue Channel");
# plt.imshow(img)
img = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(img[:,:,::-1], cmap="gray");plt.title("Merged");
```

# HSV 
show the hsv in gray, let see what happen.

```python
img = cv2.imread("New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(hsv_img)
# adapter the shot, log begin loger
# 我靠，这里好像优先级不高啊。
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray");plt.title("H Channel");
plt.subplot(142);plt.imshow(s, cmap="gray");plt.title("S Channel");
plt.subplot(143);plt.imshow(v, cmap="gray");plt.title("V Channel");
# plt.imshow(img)
img = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(img[:,:,::-1]);plt.title("Origin");
```

## turn H(color) + 10

```python
img = cv2.imread("New_Zealand_Lake.jpg", cv2.IMREAD_COLOR)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(hsv_img)
h = h + 10
img = cv2.cvtColor(cv2.merge((h, s, v)), cv2.COLOR_HSV2BGR)
# adapter the shot, log begin loger
# 我靠，这里好像优先级不高啊。
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h, cmap="gray");plt.title("H Channel");
plt.subplot(142);plt.imshow(s, cmap="gray");plt.title("S Channel");
plt.subplot(143);plt.imshow(v, cmap="gray");plt.title("V Channel");
# plt.imshow(img)
# img = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(img[:,:,::-1]);plt.title("Origin");
```
